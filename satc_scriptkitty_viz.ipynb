```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data18 = pd.read_csv('c2k1800_out.csv')
data18.head()
```

```python
data20 = pd.read_csv('c2k2000_out.csv')
data20.head()
```

```python
data25 = pd.read_csv('c2k2500_out.csv')
data25.head()
```

```python
data30 = pd.read_csv('c2k3000_out.csv')
data30.head()
```

```python
netmap = pd.read_csv('netmap.csv')
netmap.head()
```

```python
merged18 = pd.merge(data18,netmap,left_on='id',right_on='id')
merged20 = pd.merge(data20,netmap,left_on='id',right_on='id')
merged25 = pd.merge(data25,netmap,left_on='id',right_on='id')
merged30 = pd.merge(data30,netmap,left_on='id',right_on='id')
```

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,10), sharex=True, sharey=True)
sns.violinplot(x='dos', y='p', data=merged18.loc[(merged18.up_id==201)|(merged18.up_id==202)], hue='up_id', split=True, ax=ax1)
sns.violinplot(x='dos', y='p', data=merged20.loc[(merged20.up_id==201)|(merged20.up_id==202)], hue='up_id', split=True, ax=ax2)
sns.violinplot(x='dos', y='p', data=merged25.loc[(merged25.up_id==201)|(merged25.up_id==202)], hue='up_id', split=True, ax=ax3)
sns.violinplot(x='dos', y='p', data=merged30.loc[(merged30.up_id==201)|(merged30.up_id==202)], hue='up_id', split=True, ax=ax4)
ax1.set(title='1800', xlabel='', ylabel='Price')
ax2.set(title='2000', xlabel='', ylabel='')
ax3.set(title='2500', xlabel='DOS Level', ylabel='Price')
ax4.set(title='3000', xlabel='DOS Level', ylabel='')
ax1.get_legend().remove()
ax2.get_legend().remove()
ax3.get_legend().remove()
ax4.get_legend().remove()
ax4.legend(loc='lower center', bbox_to_anchor=(-0.1, -0.22),
           ncol=3, fancybox=True, shadow=True, prop={'size': 15})
fig.suptitle('Node Prices: Net Supply by 201\nPopulations = 10, Transmission cost = 2',fontsize=20)
plt.savefig('DOS_violins.png')
```

```python
fig, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,10), sharex=True, sharey=True)
sns.swarmplot(x='dos', y='p', data=merged18.loc[(merged18.up_id==201)|(merged18.up_id==202)], hue='up_id', dodge=True, ax=ax1)
sns.swarmplot(x='dos', y='p', data=merged20.loc[(merged20.up_id==201)|(merged20.up_id==202)], hue='up_id', dodge=True, ax=ax2)
sns.swarmplot(x='dos', y='p', data=merged25.loc[(merged25.up_id==201)|(merged25.up_id==202)], hue='up_id', dodge=True, ax=ax3)
sns.swarmplot(x='dos', y='p', data=merged30.loc[(merged30.up_id==201)|(merged30.up_id==202)], hue='up_id', dodge=True, ax=ax4)
ax1.set(title='1800', xlabel='', ylabel='Price')
ax2.set(title='2000', xlabel='', ylabel='')
ax3.set(title='2500', xlabel='DOS Level', ylabel='Price')
ax4.set(title='3000', xlabel='DOS Level', ylabel='')
ax1.get_legend().remove()
ax2.get_legend().remove()
ax3.get_legend().remove()
ax4.get_legend().remove()
ax4.legend(loc='lower center', bbox_to_anchor=(-0.1, -0.22),
           ncol=3, fancybox=True, shadow=True, prop={'size': 15})
fig.suptitle('Node Prices: Net Supply by 201\nPopulations = 10, Transmission cost = 2',fontsize=20)
plt.savefig('p10swarm.png')
```

```python
merge_stack = merged181000.set_index(['pop','id','dos'])
```

```python
pop_id_stack = merge_stack.unstack()
pop_id_stack.head()
```

```python
pop_id_stack['p']
```

```python
dos_lvls = pop_id_stack['p'].columns.to_list()

# for loop to construct differences in price for each dos level of each id, pop combo
# notice construction of only 3 levels
for lvl in dos_lvls[1:]:
    # for indexing
    i = 0
    pop_id_stack['p','dp{level}'.format(level=lvl)] = pop_id_stack['p',lvl] - pop_id_stack['p',0]
    
```

```python
pop_id_stack.columns
```
